## convert *.ppm to *.jpg

```python
from PIL import Image
import os
```

```python
# example
img = Image.open("origin/00000.ppm")
img.save('out.jpg', "JPEG", quality=100, optimize=True, progressive=True)
```

```python
# define
input_path = 'origin/'
output_path = 'converted/images'

# get file in folder

def get_ppm_files(dir_path):
    all_content = os.listdir(dir_path)
    
    files = []
    for item in all_content:
        if item.endswith('.ppm'):
            files.append(item)
    files.sort()
    print(len(files))
    return files
    
def convert_ppm_to_jpg(input_path, filename, dir_saved):
    out_file_name = filename.replace('.ppm', '.jpg')
    if not os.path.isdir(dir_saved):
        os.mkdir(dir_saved)
    out_file_path = os.path.join(dir_saved, out_file_name)
    
    image = Image.open(os.path.join(input_path, filename))
    image.save(out_file_path, "JPEG", quality=100, optimize=True, progressive=True)
    
filenames = get_ppm_files(input_path)
for filename in filenames:
    convert_ppm_to_jpg(input_path, filename, output_path)
```

## convert *.csv to *.xml(s)

```python
from pascal_voc_writer import Writer

# example
# <file_name>;<x1;y1;x2;y2>;<class_id>
input_str = '00000.ppm;774;411;815;446;11'

a_input = input_str.split(';')

w = Writer(a_input[0], 400, 400)
w.addObject(a_input[-1], a_input[1], a_input[2], a_input[3], a_input[4])
w.save('out.xml')
```

```python
import os
from PIL import Image

# define
input_path = 'origin/gt.txt'
image_dir = 'origin'
xml_dir = 'converted/xml'
if not os.path.isdir(xml_dir):
        os.mkdir(xml_dir)

annotation_content = []
with open(input_path) as f:
    for line in f:
        annotation_content.append(line)

rows = len(annotation_content)
i = 0
while i < rows:
    line = annotation_content[i]
    a = line.split(';')

    image_path = os.path.join(image_dir, a[0])
    image = Image.open(image_path)
    w, h = image.size
    
    w = Writer(a[0], w, h)
    w.addObject(a[-1], a[1], a[2], a[3], a[4])
    
    sub_obj = []
    for j in range(i+1, rows):
        if annotation_content[j].startswith(a[0]):
            sub_obj.append(annotation_content[j])
        else:
            break
            
    print('{}-->[{}]'.format(a[0], len(sub_obj)+1))
    if len(sub_obj) > 0:
        i += len(sub_obj) + 1
        for s_line in sub_obj:
            s_a = s_line.split(';')
            w.addObject(s_a[-1], s_a[1], s_a[2], s_a[3], s_a[4])
    else:
        i += 1
    xml_name = a[0].replace('.ppm', '.xml')
    w.save(os.path.join(xml_dir, xml_name))
```
